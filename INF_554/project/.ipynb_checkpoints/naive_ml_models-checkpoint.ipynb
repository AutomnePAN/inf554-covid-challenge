## First Model: Classification, y = 0?

From the histgram and statistic value above, we found that the nealy 2/3 of the tweet have 0 retweet, so we want to start by trying to classify if the tweet has retweet or not.

```python
y1 = (y > 0)
```

```python
y1.sum()
```

```python
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X, y1, test_size=0.25, random_state=42)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=10, random_state=42).fit(X_train_1, y_train_1)
```

```python
# Importance of features
clf.feature_importances_
```

```python
prediction = clf.predict(X_train_1)
```

```python
print( "f1 score: ", f1_score(prediction, y_train_1, average="macro") )
print( "accuracy: ", accuracy_score(prediction, y_train_1) )
```

```python
for threshold in [0.15, 0.20, 0.25, 0.35, 0.40]:
    prediction = clf.predict_proba(X_test_1)
    y_pred = (prediction[:,1] > threshold)
    print(threshold)
    print( "confusion matrix: \n" )
    print( confusion_matrix(y_test_1, y_pred ) )
```

We pick the threshold 0.2
